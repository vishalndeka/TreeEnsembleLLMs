```python
import numpy as np
import matplotlib.pyplot as plt
import ast
```

# Some pre-processing

```python
def file_to_matrix(filename):
  liStr = []
  with open(filename, "r", encoding="cp1252", errors="ignore") as file:
    while True:
      line = file.readline()
      if not line:
          break
      # Process the line
      liStr.append(line)
  li = []
  for _ in liStr:
    li.append(ast.literal_eval(_))
  return li
```

```python
li = file_to_matrix('Experiment_Results/blender_scores.txt')
blender_scores = []
for i in range(len(li)):
    l = []
    for j in range(len(li[i])):
        l.append(li[i][j][0])
    blender_scores.append(l)
li = file_to_matrix('Experiment_Results/scores_tree_rerun.txt')
tree_scores = []
for i in range(len(li)):
    l = []
    for j in range(len(li[i])):
        l.append(li[i][j][0])
    tree_scores.append(l)
```

# Plotting results

```python
plt.style.use('_mpl-gallery')

x = np.linspace(0, len(blender_scores), len(blender_scores))
y = np.average(blender_scores, axis = 1)

min_y = np.min(y)
min_x = x[y.argmin()]
max_y = np.max(y)
max_x = x[y.argmax()]

# plot
fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(x, y, linewidth=2.0)
ax.plot(min_x, min_y, 'ro')
ax.plot(max_x, max_y, 'go')
ax.text(min_x, min_y, min_y)
ax.text(max_x, max_y, max_y)
plt.title('Average BERT score for Conv Blender: ' + str(np.sum(y)/len(blender_scores)))
plt.show()
```

```python
plt.style.use('_mpl-gallery')

x = np.linspace(0, len(tree_scores), len(tree_scores))
y = np.average(tree_scores, axis = 1)

min_y = np.min(y)
min_x = x[y.argmin()]
max_y = np.max(y)
max_x = x[y.argmax()]

# plot
fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(x, y, linewidth=2.0)
ax.plot(min_x, min_y, 'ro')
ax.plot(max_x, max_y, 'go')
ax.text(min_x, min_y, min_y)
ax.text(max_x, max_y, max_y)
plt.title('Average BERT score for Tree Structure Ensemble: ' + str(np.sum(y)/len(tree_scores)))
plt.show()
```
